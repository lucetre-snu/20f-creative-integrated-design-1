```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy.random import randn

loss = "bce wbce(237) wbce(59) wbce(15) f1".split()
plt.rcParams["figure.figsize"] = (16,4)
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.color'] = 'r'
plt.rcParams['axes.grid'] = True
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False

for l in loss:
    # epoch/loss/acc/f1/tp/tn/fp/fn
    df = pd.read_csv(f"{l}/val.tsv", delimiter='\t', header=None, skiprows=1)
    val = df[1].to_numpy()
    df = pd.read_csv(f"{l}/train.tsv", delimiter='\t', header=None, skiprows=1)
    train = df[1].to_numpy()
#     plt.suptitle('Loss Graph', fontsize=20)
    plt.plot(val)
    plt.plot(train)
    plt.legend(['val', 'train'], fontsize=15)
    plt.xlabel('epoch', fontsize=15)
    plt.ylabel('loss', fontsize=15)
    plt.savefig(f'loss-{l}.png', dpi=300)
    plt.show()
    
```
